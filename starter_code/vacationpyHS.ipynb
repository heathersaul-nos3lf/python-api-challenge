### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
locations = weather_data[['Lat','Lon']].astype(float)
weights = weather_data['Humidity'].astype(float)
fig = gmaps.figure(center=(20.0,0), zoom_level=2)
fig.add_layer(gmaps.heatmap_layer(locations, weights=weights))
fig
heatmap_layer = gmaps.heatmap_layer(
    locations, weights=weights, dissipating=False, max_intensity=100, point_radius=1)
fig.add_layer(heatmap_layer)
fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
url = "https://maps.googleapis.com/maps/api/place/textsearch/json?"
hotels=[]
for i in range(0,len(weath_dat)):
    query = 'Lodging in '+list(weath_dat['City'])[i]

    r = requests.get(url + 'query=' + query +
                        '&key=' + g_key) 
    x= r.json()['results']
    hotels.append(x[0]['name'])
```
